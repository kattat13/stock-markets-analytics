# Question 4. [Stocks OHLCV] 52-weeks range ratio (2023) for the selected stocks

## Find the largest range ratio [=(max-min)/max] of Adj.Close prices in 2023

```python
# Download the 2023 daily OHLCV data on Yahoo Finance
# for top6 stocks on earnings (https://companiesmarketcap.com/most-profitable-companies/):
# 2222.SR, BRK-B, AAPL, MSFT, GOOG, JPM

top6_companies = ['2222.SR', 'BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM']

top6 = yf.download(tickers = top6_companies,
                   start = '2023-01-01',
                   end = '2023-12-31')['Adj Close']
top6.tail()
```

```python
# Calculate maximum-minimim "Adj.Close" price for each
# stock and divide it by the maximum "Adj.Close" value.
# Round the result to two decimal places.

for stock_name in top6.columns:
    range_ratio = (top6[stock_name].max() - top6[stock_name].min()) / top6[stock_name].max()
    print(f'{stock_name}\'s range ratio: {range_ratio:.2f}')
```

# Question 5. [Stocks] Dividend Yield

## Find the largest dividend yield for the same set of stocks

```python
dividends_sum = []
```

```python
for name in top6_companies:
    x = yf.Ticker(name).get_dividends()
    dividends_sum.append(x[(x.index >= '2023-01-01') & (x.index < '2024-01-01')].sum())
```

```python
dividends_sum
```

```python
last_two_rows = top6.tail(2).copy()
last_two_rows
```

```python
last_two_rows.fillna(method='ffill', inplace=True)
last_two_rows
```

```python
dividends_sum / last_two_rows.iloc[-1]
```
